```python
def add_history_month(x):
    for index, item in x.items():
        if pd.isna(item):
            x[index] = x.shift(1)[index] + 1
    return x
df.Credit_History_Age = df.groupby('Customer_ID').Credit_History_Age.apply(lambda x: add_history_month(x))
```

```python
def add_history_month(x):
    for index, item in x.items():
        if pd.isna(item):
            x[index] = x.shift(-1)[index] - 1
    return x
df.Credit_History_Age = df.groupby('Customer_ID').Credit_History_Age.apply(lambda x: add_history_month(x))
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df['Credit_Score'] = df['Credit_Score'].astype('str')
df.replace({'Credit_Score' : {'0.0' : 'Poor', '0.5' : 'Standard', '1.0' : 'Good'}}, inplace=True)
```

```python
X = df.drop(['Credit_Score'], axis = 1)
y = df['Credit_Score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
print(accuracy_score(y_test,pred))
```

```python
error_rate = []

for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,30), error_rate, marker='o', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.grid()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
print(classification_report(y_test, y_pred))
```
